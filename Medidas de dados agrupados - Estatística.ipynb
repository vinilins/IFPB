```python
def dist_frequencia_classes(x, nc='sturges'):
    fi,classes= np.histogram(x, bins=nc) 
    fr = np.around(fi*100/len(x),decimals=2)
    Fc = fi.cumsum()
    classe_str = []
    for i in range(len(classes)):
        if i<=len(classes)-2:
            classe_str.append('{:2.2f} |-- {:2.2f}' .format(classes[i],classes[i+1]))
        else:
            break        
    return pd.DataFrame({'Classes':classe_str, 
                         'Fi': fi,
                         'Fr': fr,
                         'Fc': Fc
                        }).set_index('Classes')

def medidasDeDadosAgrupados(distriFreq):
    maiorValor = 0
    for i in distriFreq['Fi']:
        if i>maiorValor:
            maiorValor = i
    moda = distriFreq['Fi'] == maiorValor
    classesBaguncadas = []
    cont = 0
    for i in moda:
        if i:
            classesBaguncadas.append(moda.index[cont])
        cont+=1
    classesCertas = []
    for classe in classesBaguncadas:
        controle = 0
        listaAuxiliar = []
        num1 = ''
        num2 = ''
        for letra in classe:
            if (letra.isdigit() or letra=='.') and controle==0:
                num1+=letra
            elif letra=='|':
                controle = 1
            elif (letra.isdigit() or letra=='.') and controle==1:
                num2+=letra
        listaAuxiliar.append(num1)
        listaAuxiliar.append(num2)
        classesCertas.append(listaAuxiliar)
    modaFinal = []
    for classe in classesCertas:
        num1 = float(classe[0])
        num2 = float(classe[1])
        modaFinal.append((num1+num2)/2)
    return modaFinal
```

```python
moda = medidasDeDadosAgrupados(distriFrequenciaFare)
moda
```
